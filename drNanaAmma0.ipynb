```python
fig0, axs = plt.subplots(ncols=1, nrows=20, figsize=(20, 30), sharex=True)

for row, colname in enumerate(data00.columns):
    axs[row, ].plot(data00[colname], linewidth=2.5, alpha=0.5, label=colname)
    axs[row, ].grid()
    axs[row, ].set_xticks([])
    axs[row, ].set_yticks([0, 5])
    axs[row,].legend()
    axs[row,].margins(x=0)
    axs[row,]
    
fig0.supylabel('Price')
fig0.supxlabel('Dates')
plt.xticks(data00.index.to_list(),  rotation=90)
plt.suptitle('Rendering all data and missing data')
plt.show()
```

```python
fig0, axs = plt.subplots(ncols=1, nrows=20, figsize=(20, 30), sharex=True)

pad = 0.02
for row, colname in enumerate(data00.columns):
    missing_data_mask = pd.isna(data00[colname])
    missing_value_placeholder = data00[colname].max() + pad
    data00_na_replaced = data00[colname].fillna(missing_value_placeholder)
    axs[row, ].scatter(data00.index.to_list(), data00[colname], marker='.', s=40,  label=colname)
    axs[row, ].scatter(data00.index[missing_data_mask], data00_na_replaced[missing_data_mask], marker='.', s=40, label="Miss'g data")

    axs[row, ].grid()
    axs[row, ].set_xticks([])
    axs[row, ].set_yticks([0, 5])
    axs[row,].legend()
    axs[row,].margins(x=0)
    axs[row,]
    
fig0.supylabel('Price', fontsize=20)
fig0.supxlabel('Dates', fontsize=20)
plt.xticks(data00.index.to_list(),  rotation=90)
plt.suptitle('Rendering all data and missing data')
plt.show()
# plot nan as red, in plot
# plot imputed data for a few of the cols
# create masks, call scatter to plot point for which the mask is ture
```
